```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv('C:/Users/ML Lab/Desktop/122B1F003/Employee_Salary_Dataset.csv')
```

```python
# Display first few rows
print("First five rows:")
print(df.head())
```

```python
df.info()
```

```python
print("\nMissing values:")
print(df.isnull().sum())
```

```python
df.fillna({'Salary': df['Salary'].median()}, inplace=True)
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Encoding categorical variables (if any exist)
categorical_cols = df.select_dtypes(include=['object']).columns
if len(categorical_cols) > 0:
    df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
    print("\nCategorical columns encoded:", categorical_cols.tolist())
```

```python
# Correlation Matrix
correlation_matrix = df.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
# Heatmap of correlation
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Boxplot for Salary
plt.figure(figsize=(6, 3))
sns.boxplot(x=df['Salary'])
plt.title("Salary Boxplot")
plt.show()
```

```python
# Histogram of Salary
plt.figure(figsize=(6, 3))
sns.histplot(df['Salary'], bins=30, kde=True)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```
